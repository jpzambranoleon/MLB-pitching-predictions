```python
import requests
import pandas as pd
import json
```

```python
url = "https://api.sportradar.com/mlb/trial/v7/en/games/b1f0e4d6-e5b6-4477-a4f8-5f6695758267/pbp.json?api_key=S09vsqYLzQadSsAMNuqNV8ZC5pabMVOPA8oZCZva"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

play_by_play = response.json()
```

```python
# Check if the request was successful
if response.status_code == 200:
    # Parse and display the pitch metrics data
    print(json.dumps(play_by_play, indent=1, ensure_ascii=False))
else:
    print("Error:", response.status_code)
```

```python
new_json = {
    "game_id": play_by_play["game"]["id"],
    "innings": play_by_play["game"]["innings"]

}

new_json
```

```python
df = pd.json_normalize(new_json)
df = df.explode("innings").reset_index()
df = df.drop("index", axis=1)
df
```

```python
innings_parse_df = df["innings"].apply(pd.Series)
innings_parse_df = innings_parse_df.drop('scoring', axis=1)
innings_parse_df
```

```python
merged_df_1 = pd.concat([df.drop("innings", axis=1), innings_parse_df], axis=1)
merged_df_1
```

```python
merged_df_1 = merged_df_1.explode("halfs").reset_index()
merged_df_1
```

```python
halfs_parse_df = merged_df_1["halfs"].apply(pd.Series)
halfs_parse_df = halfs_parse_df.drop('weather', axis=1)
halfs_parse_df
```

```python
merged_df_2 = pd.concat([merged_df_1.drop("halfs", axis=1), halfs_parse_df], axis=1)
merged_df_2 = merged_df_2.dropna()
merged_df_2 = merged_df_2.drop(merged_df_2[merged_df_2["sequence"] == 0].index)
merged_df_2
```

```python
merged_df_2 = merged_df_2.explode("events").reset_index()
merged_df_2
```

```python
events_parse_df = merged_df_2["events"].apply(pd.Series)
events_parse_df = events_parse_df.drop(["lineup", 0], axis=1)
events_parse_df
```

```python
merged_df_3 = pd.concat([merged_df_2.drop("events", axis=1), events_parse_df], axis=1)
merged_df_3
```

```python
at_bat_parse_df = merged_df_3["at_bat"].apply(pd.Series)
at_bat_parse_df
```

```python
merged_df_4 = pd.concat([merged_df_3.drop(["at_bat"], axis=1), at_bat_parse_df], axis=1)
merged_df_4 = merged_df_4.drop(['level_0', 'index'], axis=1)
merged_df_4
```
